# Complex Analysis, Priestly Question 1.8

Evaluate $\sum_{k=1}^n e^{ik\theta}$.  Deduce that:

$$-1 + 2\sum_{k=1}^n \cos k\theta = \frac{\sin \left(n + \frac{1}{2}\right)n}{\sin \frac{1}{2} \theta} \text{ for } \theta \neq 2m\pi ( m\in \mathbb{Z} )\;.$$

Find a similar expression for $\sum_{k=1}^n \sin k\theta$.

Solution:

When $\theta \neq 2m\pi$, $\sin \frac{1}{2}\theta \neq 0$ so there is no division by 0.

Recall the geometric series formula:
$$\sum_{k=0}^n r^k = \frac{1-r^{n+1}}{1-r}\;,$$
and let $r = e^{i\theta}$:
$$\sum_{k=0}^n e^{i\theta k} = \frac{1-e^{i(n+1)\theta}}{1-e^{i\theta}}\;,$$

Since 

\begin{align*}
e^{i\theta} &= \cos k\theta + i\sin \theta\\
e^{-i\theta} &=  \cos k\theta -i\sin \theta\;,\\
\end{align*}

we can substitute these expressions in to the geometric series formula:
\begin{align*}
\sum_{k=0}^n e^{ik\theta} =\frac{1-\cos (n+1) \theta +i \sin (n+1)\theta}{1-\cos\theta - i \sin \theta} \\
\sum_{k=0}^n e^{-ik\theta} =\frac{1-\cos (n+1) \theta - i \sin (n+1)\theta}{1-\cos\theta + i \sin \theta} \\
\end{align*}

In addition, we have:
\begin{align*}
\sum_{k=0}^n e^{ik\theta} &= \sum_{k=0}^n \left( \cos k\theta + i\sin k\theta \right)\\
 \sum_{k=0}^n e^{-ik\theta} &=  \sum_{k=0}^n \left( \cos k\theta -i\sin k\theta \right)\;.\\
\end{align*}

Therefore, we can get a summation for cosine by adding the above summations:

$$ 2\sum_{k=0}^n \cos k\theta = \sum_{k=0}^n e^{ik\theta} +  \sum_{k=0}^n e^{-ik\theta} $$

Before adding the two summations, we first simplify them further.  The summation $\sum_{k=0}^n e^{ik\theta}$ simplifies as follows:

\begin{align*}
\sum_{k=0}^n e^{ik\theta} &=\frac{1-\cos (n+1) \theta - i \sin (n+1)\theta}{1-\cos\theta - i \sin \theta} \\
&=\frac{1-\cos (n+1) \theta - i \sin (n+1)\theta}{1-\cos\theta - i \sin \theta} \times \frac{1-\cos \theta + i \sin \theta}{1-\cos \theta + i \sin \theta} \\
&= \frac{(1-\cos(n+1)\theta)(1-\cos \theta) + \sin(n+1)\theta\sin\theta}{(1-\cos \theta)^2 + \sin^2\theta} + i\frac{\sin\theta(1-\cos(n+1)\theta) -\sin(n+1)\theta(1-\cos\theta)}{(1-\cos \theta)^2 + \sin^2\theta}
\end{align*}

The summation $\sum_{k=0}^n e^{-ik\theta}$ simplify as follows:

\begin{align*}
\sum_{k=0}^n e^{-ik\theta} &=\frac{1-\cos (n+1) \theta - i \sin (n+1)\theta}{1-\cos\theta + i \sin \theta}\times \frac{1-\cos \theta- i \sin \theta}{1-\cos \theta - i \sin \theta} \\
&=\frac{(1-\cos(n+1)\theta)(1-\cos\theta)+\sin(n+1)\theta \sin\theta}{(1-\cos \theta)^2 + \sin^2\theta} - i\frac{\sin\theta(1-\cos(n+1)\theta) -\sin(n+1)\theta(1-\cos\theta)}{(1-\cos \theta)^2 + \sin^2\theta}
\end{align*}

Now we have the expression for $\sum_{k=0}^n \cos k\theta$:

$$2\sum_{k=0}^n \cos k\theta = 2\frac{(1-\cos(n+1)\theta)(1-\cos\theta)+\sin(n+1)\theta \sin\theta}{(1-\cos \theta)^2  + \sin^2\theta}\;.$$

The denominator can be further simplified:
\begin{align*}
&(1-\cos \theta)^2  + \sin^2\theta \\
&=1-2\cos\theta + \cos^2 \theta + \sin^2\theta \\
&= 2(1-\cos \theta)\;.
\end{align*}

The following trigonometric identities will help us to further simplify this expression:

\begin{align*}
\sin(u \pm v) &= \sin u \cos v \pm \cos u \sin v \;, \\
\cos(u \pm v) &= \cos u \cos v \mp \sin u \sin v \;, \\
\sin 2u &= 2 \sin u \cos u \;,\\
\sin^2 u &= \frac{1-cos 2u}{2}\;,
\end{align*}

Using all of the above equations, the simplification proceeds as follows:

\begin{align*}
2\sum_{k=0}^n \cos k\theta &= 2\frac{(1-\cos(n+1)\theta)(1-\cos\theta)+\sin(n+1)\theta \sin\theta}{(1-\cos \theta)^2  + \sin^2\theta} \\
&= \frac{ (1-\cos(n+1)\theta)(1-\cos\theta)+\sin(n+1)\theta \sin\theta)}{2(1-\cos\theta)} \\
&= \frac{ 1-\cos\theta-\cos(n+1)\theta + \cos(n+1)\theta \cos\theta+\sin(n+1)\theta \sin\theta}{1-\cos\theta}\\
&= \frac{ 1-\cos\theta-\cos(n+1)\theta+ \cos(n+1)\theta \cos\theta+\sin(n+1)\theta \sin\theta}{1-\cos\theta}\\
&= \frac{ 1-\cos\theta-\cos(n+1)\theta+ \cos n\theta }{1-\cos\theta}\\
&= 1-\frac{ \cos(n+1)\theta- \cos n\theta }{1-\cos\theta}\\
&= 1-\frac{ \cos n\theta \cos \theta - \sin n\theta\sin\theta- \cos n\theta }{1-\cos\theta}\\
&= 1+\frac{ \cos n\theta - \cos n\theta \cos \theta +\sin n\theta\sin\theta}{1-\cos\theta}\\
&= 1+\frac{ \cos n\theta(1 - \cos \theta) + \sin n\theta\sin\theta}{1-\cos\theta}\\
&= 1+\frac{ \cos n\theta (2\sin^2\frac{\theta}{2}) + \sin n\theta(2\sin\frac{\theta}{2}\cos\frac{\theta}{2})}{2\sin^2\frac{\theta}{2}}\\
&= 1+\frac{ \cos n\theta \sin\frac{\theta}{2} + \sin n\theta \cos\frac{\theta}{2}}{\sin\frac{\theta}{2}}\\
&= 1+\frac{ \sin \left(n+\frac{1}{2}\right) \theta}{\sin\frac{\theta}{2}}\\
\end{align*}

Moving the 1 to the left hand side concludes the simplification:
$$-1 + 2\sum_{k=0}^n \cos k\theta = \frac{ \sin \left(n+\frac{1}{2}\right) \theta}{\sin\frac{\theta}{2}} $$
